```python
from math import pi
import numpy as np
from scipy import interpolate
import matplotlib.pyplot as plt
```

```python
f = lambda x: 1/np.cos(x) 
seg = [0, pi/3]
n = 12 #кількість вузлів
X = []
Y = []
c = seg[0] + seg[1] / n
for i in range(n):
    X.append(seg[0] + i * c)
    Y.append(f(X[i]))

print(X)
print(Y)
```

```python
poly = interpolate.lagrange(X, Y)
print(f"Lagrange polynomial coefs:\n{poly.coeffs}")

X_1 = np.linspace(0, pi/3, 1000)
Y_1 = 1 / np.cos(X_1)
```

```python
plt.scatter(X, Y, label='interpolation nodes', c = "olivedrab")
plt.plot(X_1, Y_1, label='f(x)', linestyle='-.', c = "mediumpurple")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
plt.plot(np.abs(poly(X_1) - Y_1)/n,'-')
plt.title("Lagrange error")
plt.show()
```

```python
spline = interpolate.CubicSpline(X, Y)

plt.scatter(X, Y, label='interpolation nodes', c = "olivedrab")
plt.plot(X_1, spline(X_1), label='spline', linestyle='-.', c = "mediumpurple")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
plt.plot(np.abs(spline(X_1) - Y_1)/n,'-')
plt.title("Spline error")
plt.show()
```
